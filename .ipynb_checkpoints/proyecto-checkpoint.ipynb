```python
import pandas as pd
#comentario sebas
```

```python
# Leer el archivo CSV
archivo_csv = 'energia.csv'
datos_csv = pd.read_csv(archivo_csv)
```

```python
# Mostrar información del CSV
print("Contenido del CSV:")
print(datos_csv.head())
```

```python
datos_csv.info()
```

```python
# Estadísticas descriptivas de las columnas numéricas
print("\nEstadísticas descriptivas:")
print(datos_csv.describe())
```

```python
# Verificar si hay valores nulos
print("\nValores nulos por columna:")
print(datos_csv.isnull().sum())
```

```python
# get all the unique values in the 'PROVINCE' column
province = datos_csv['DEPARTAMENTO'].unique()
city = datos_csv['MUNICIPIO'].unique()
zone = datos_csv['LOCALIDAD'].unique()

# sort them alphabetically and then take a closer look
province.sort()
city.sort()
zone.sort()
print(f'DEPARTAMENTO: {province}')
print(f'MUNICIPIO: {city}')
print(f'LOCALIDAD: {zone}')
```
